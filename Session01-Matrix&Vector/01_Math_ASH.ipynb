## 1.2 Unit Vectors

```python
def unit_vector(v):
    norm = np.linalg.norm(v)
    return v/norm
```

```python
v = np.array([3, -4, 5])
u = unit_vector(v)
print("Unit Vector is ", u)
```

## 1.7 Angle Between Two Vectors and Its Applications in Machine Learning

```python
def cosin_sim(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)
```

```python
a = np.array([3, -5])
b = np.array([7, 1])
```

```python
cos_theta = cosin_sim(a, b)
```

```python
np.degrees(np.arccos(cos_theta))
```
