```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
df = pd.read_csv("houses.csv")
df.head()
```

```python
df.info()
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
X_train
```

```python
scaler = StandardScaler()
```

```python
X_train_s = scaler.fit_transform(X_train)
X_train_s
```

```python
X_test_s = scaler.transform(X_test)
X_test_s
```

```python
LR = LinearRegression()
LR.fit(X_train_s, y_train)
```

```python
y_pred = LR.predict(X_test_s)
y_pred
```

```python
MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)

print(f"MAE: {MAE}, MSE: {MSE}, RMSE: {RMSE}")
```

```python
def cost_function(x, y, W, b):
    m = x.shape[0]
    f_wb = np.dot(x, W) + b
    error = f_wb - y
    J = (1 / (m)) * np.sum(error ** 2)
    return J, error
```

```python
def gradient_descent(alpha, x, y, n_iter=10000, tolerance=1e-6):
    m, n = x.shape
    W = np.zeros(n)
    b = 0
    J_hist = []

    for iter in range(n_iter):
        J, error = cost_function(x, y, W, b)
        J_hist.append(J)

        # Stop if improvement is below tolerance
        if iter > 0 and abs(J_hist[-1] - J_hist[-2]) < tolerance:
            break
        
        # Compute Gradients
        dj_dw = (1 / m) * np.dot(x.T, error)
        dj_db = (1 / m) * np.sum(error)

        # Update Parameters
        W -= alpha * dj_dw
        b -= alpha * dj_db

    return W, b, J_hist
```

```python
alpha = 0.01
W, b, J_hist = gradient_descent(alpha, X_train_s, y_train)
```

```python
y_prediction = np.dot(X_test_s, W) + b
```

```python
y_prediction
```

```python
# Scatter plot of true vs predicted values
plt.scatter(y_test, y_prediction, alpha=0.6, label="Predicted vs True")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Linear Regression Prediction")
plt.legend()
plt.show()
```

```python
# Plot Cost Function Convergence
plt.plot(J_hist, color="black")
plt.xlabel("Iteration")
plt.ylabel("J(W, b)")
plt.title("Gradient Descent Cost Function Convergence")
plt.show()
```
