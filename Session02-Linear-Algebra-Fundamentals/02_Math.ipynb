```python
import numpy as np
```

## Matrix Multiplication

```python
def matrix_multiply(A, B):

    # Check if multiplication is possible
    if len(A[0]) != len(B):
        raise ValueError("Number of columns in A must equal number of rows in B")
    
    # Initialize the resulting matrix with zeros
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    
    # Compute matrix multiplication
    for i in range(len(A)):  # Loop over rows of A
        for j in range(len(B[0])):  # Loop over columns of B
            for k in range(len(B)):  # Loop over common dimension
                result[i][j] += A[i][k] * B[k][j]
    
    return result

# Example Matrices
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

# Perform Matrix Multiplication
result = matrix_multiply(A, B)

# Display Result
print("Matrix A:")
for row in A:
    print(row)

print("\nMatrix B:")
for row in B:
    print(row)

print("\nResulting Matrix C:")
for row in result:
    print(row)
```

```python
# Example Matrices
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

# Perform Matrix Multiplication
C = np.dot(A, B)

# Display Results
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nResulting Matrix C:")
print(C)
```

## Hadamard Product

```python
# Example Matrices
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8, 9],
    [10, 11, 12]
])

# Perform Hadamard Product
C = np.multiply(A, B)  # Element-wise multiplication
# Alternatively, you can use `C = A * B` for the same result

# Display Results
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nHadamard Product (Element-wise multiplication):")
print(C)
```

## Outer Product

```python
# Example vectors
u = np.array([1, 2, 3])  # Vector of size 3
v = np.array([4, 5])     # Vector of size 2

# Perform Outer Product
result = np.outer(u, v)

# Display Results
print("Vector u:", u)
print("Vector v:", v)
print("\nOuter Product (NumPy Implementation):")
print(result)
```

## Inner Product

```python
# Example Vectors
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

# Compute Inner Product
result = np.dot(u, v)

# Display Result
print(f"Vector u: {u}")
print(f"Vector v: {v}")
print(f"Inner Product: {result}")
```

## Dot Product

```python
# Example Vectors
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

# Compute Dot Product
dot_product = np.dot(u, v)

print("Vector u:", u)
print("Vector v:", v)
print("Dot Product:", dot_product)
```

## Kronecker Product

```python
# Define Matrices
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[0, 5],
              [6, 7]])

# Compute Kronecker Product
kronecker_product = np.kron(A, B)

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nKronecker Product (A ⊗ B):")
print(kronecker_product)
```

## Cross Product

```python
u = np.array([2, 3, 4])
v = np.array([5, 6, 7])
result = np.cross(u, v)
print("Cross Product:", result)
```
