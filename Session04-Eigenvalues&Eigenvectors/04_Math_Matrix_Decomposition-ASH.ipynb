# Import

```python
import numpy as np
from scipy.linalg import lu
from scipy.linalg import lu_factor, lu_solve
```

# LU Decomposition

```python
A = np.array([[2, 1, 1], [4, -6, 0], [-2, 7, 2]])
b = np.array([5, -2, 9])
```

```python
lu, piv = lu_factor(A)
```

```python
lu
```

```python
piv
```

```python
x = lu_solve((lu, piv), b)
```

# Eigenvalue and Eigenvector Decomposition

```python
A = np.array([[4, 1], [2, 3]]) # a11, a22
```

```python
# det(A - lambda I) = 0 -> +1 * lambda^2 - trace(A)lambda + det(A) = 0
# trace(A) = Sum of diagnal elements
```

```python
trace_A = np.trace(A)
trace_A
```

```python
det_A = np.linalg.det(A)
det_A
```

```python
# a x^2 + b x + c, x (-b +/- sqrt(b62 - 4ac) )/ 2a
# lambda = -b +/- sqrt(b^2 - 4ac)]/2a
# a=1, b = -trace(A), c = det(A)
```

```python
b = -trace_A
c = det_A
```

```python
eigenvalue1 = (-b + np.sqrt(b**2 - 4 * 1 * c)) / (2 * 1)
eigenvalue2 = (-b - np.sqrt(b**2 - 4 * 1 * c)) / (2 * 1)
```

```python
eigenvalue1, eigenvalue2
```

```python
# (A - lambda I) = 0

lambda_1 = eigenvalue1
I = np.eye(2)
A_lambda_1 = A - lambda_1 * I
A_lambda_1
```

```python
_, _, v1 = np.linalg.svd(A_lambda_1)
v1
```

```python
np.linalg.eig(A)
```
