## Section 1

```python
import numpy as np
```

```python
v1 = np.array([1, 2])
v2 = np.array([3, 4])
```

```python
v1 + v2
```

```python
3 * v1
```

```python
def is_in_subspace(v, subspace_vectores):
    A = np.column_stack(subspace_vectores)
    try:
        coeffs = np.linalg.lstsq(A, v, rcond=None)[0]
        return np.allclose(np.dot(A, coeffs), v)
    except np.linalg.LinAlgError:
        return False
```

```python
v = np.array([5, 7])
subspace = [np.array([1, 0]), np.array([0, 1])]
is_in_subspace(v, subspace)
```

## Section 2

```python
def linear_comb(vectors, scalars):
    result = np.zeros_like(vectors[0])
    for v, s in zip(vectors, scalars):
        result += s * v
    return result
```

```python
v1, v2
```

```python
scalars = [2, -1]
```

```python
linear_comb([v1, v2], scalars)
```

## Section 3

```python
def are_linearly_independent(vectors):
    A = np.column_stack(vectors)
    return np.linalg.matrix_rank(A) == len(vectors)
```

```python
v1 = np.array([1, 2])
v2 = np.array([2, 4])
v3 = np.array([3, 1])
```

```python
are_linearly_independent([v1, v2])
```

```python
are_linearly_independent([v1, v3])
```
